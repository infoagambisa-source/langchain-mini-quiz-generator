**STEP 1: Set Up Environment**

Set up the development environment in three parts:
 -- Install LangChain packages.
 -- Import the required modules for the model, prompts, chains, and API access.
 -- Configure the API key as an environment variable.

```python
#Importing Libraries

from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

from dotenv import load_dotenv
```

```python
#API Key Configuration
import os

load_dotenv() # Loads variable from .env
api_key = os.getenv("OPENAI_API_KEY")


print("✓ API Key configured successfully")
print("(Key is stored securely and not displayed)")
```

**STEP 2: Initialize Language Model**

```python
#Initialize the OpenAI LLM with optimized parameters
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7,
    api_key=api_key
)

print("✓ OpenAI LLM initialized successfully")
```

**STEP 3: Design Prompt Templates**

```python
# Chain 1 - Generate a question
#This template takes a topic and generates an appropriate beginner question
question_prompt = PromptTemplate(
    input_variables=['topic'],
    template="""
    Generate a beginner-level quiz question about {topi}.
    Make the question clear, simple and suitable for someone new to the topic.
    """
)


# Chain 2 - Answer the question
# This template takes the generated question and produces an answer with explanation
answer_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
    Answer the following question in a clear and detailed way:
    Question: {question}

    Provide:
    1. The correct answer
    2. A short explanation of why it is correct

    Format the response clearly.
    """
)
```
